```python
import pandas as pd
import numpy as np
```

```python
iris=pd.read_csv("Iris.csv")
iris
```

```python
X=iris.drop(['Species'], axis=1)
```

```python
X
```

```python
Y=iris['Species']
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler #adjusting values measured on different scales to a common scale
scaler=MinMaxScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
Y_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
Y_Pred = gnb.predict(X_test) 
```

```python
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score, confusion_matrix
accuracy_nb=round(accuracy_score(Y_test,Y_Pred)* 100, 2)
acc_gaussian = round(gnb.score(X_train, Y_train) * 100, 2)
cm = confusion_matrix(Y_test, Y_Pred)
accuracy = accuracy_score(Y_test,Y_Pred)
precision =precision_score(Y_test, Y_Pred,average='micro')
recall =  recall_score(Y_test, Y_Pred,average='micro')
f1 = f1_score(Y_test,Y_Pred,average='micro')
```

```python
print("Accuracy GNB:",accuracy)
print("Precision GNB:",precision)
print("Recall GNB:",recall)
print("F1-Score GNB:",f1)

print(cm)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=gnb.classes_)
disp.plot()
```
